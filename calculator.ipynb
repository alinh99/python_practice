```python
def add(n1, n2):
    result = n1 + n2
    return result
def multiply(n1, n2):
    result = n1 * n2
    return result
def subtract(n1, n2):
    result = n1 - n2
    return result
def division(n1, n2):
    if n2 == 0:
        return 'Invalid input in the second number'
    else:
        return n1 / n2

logo = """
 _____________________
|  _________________  |
| | Pythonista   0. | |  .----------------.  .----------------.  .----------------.  .----------------. 
| |_________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___   ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 | | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___| |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 | | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___| |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 | | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___| |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = | | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___| |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|_____________________|
"""
print(logo)

operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': division
}
def calculator():
    is_continue = True
    num1 = float(input("What is the first number? "))
    for operation in operations:
        print(operation)
    while is_continue:
        operation_symbol = input('Pick an operation: ')
        num2 = float(input("What is the next number? "))
        answers = operations[operation_symbol]
        answer = answers(num1, num2)
        print(f"{num1} {operation_symbol} {num2} = {answer}")
        calculate_continue = input(f"Type 'y' to continue calculating with {answer}. Type 'r' to restart calculator. Type 'n' to exit.: ")
        if calculate_continue == 'y':
            num1 = answer
        elif calculate_continue == 'n':
            is_continue = False
            print('Goodbye.')
        elif calculate_continue == 'r':
            is_continue = False
            calculator()
        else:
            is_continue = False
            print('Invalid input.')
calculator()
```
