```python
import random
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
logo = """
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
"""

def deal_card(card):
    return random.choice(card)

def calculate_score(card_list):
    score = 0
    for item in card_list:
        score += item
    return score

def check_winner(user, computer):
    computer_score = calculate_score(computer)
    user_score = calculate_score(user)
    if user_score > 20 or computer_score > 20:
        if 11 in user or 11 in computer:
            cards[0] = 1
    if user_score > 21:
        return 'You Lose.'
    if 11 and 10 in user:
        user_score = 0

    if 11 and 10 in computer:
        computer_score = 0

    if user_score == 0:
        return "You Win."
    if computer_score == 0:
        return "You Lose."

    if user_score < 15:
        return 'You Lose'
    elif user_score == computer_score:
        return "You Draw"
    elif user_score < computer_score:
        return "You Lose."
    elif user_score > computer_score:
        return "You Win."

should_continue = True
play_game = input("Do you want to play a game of Blackjack. Type 'y' or 'n': ")
while should_continue:
    if play_game == 'y':
        user_cards = []
        computer_cards = []
        print(logo)
        user_cards.append(deal_card(cards))
        user_cards.append(deal_card(cards))
        user_score = calculate_score(user_cards)
        computer_cards.append(deal_card(cards))
        computer_cards.append(deal_card(cards))
        print(computer_cards)
        computer_score = calculate_score(computer_cards)
        print(f"Your cards: {user_cards}, current score: {user_score}")
        print(f"Computer's first card: {computer_cards[0]}")
        while user_score < 21:
            add_card = input("Type 'y' to get another card. Type 'n' to pass.")
            if add_card == 'y':
                user_cards.append(deal_card(cards))
                user_score = calculate_score(user_cards)
                print(f"user_cards: {user_cards}")
                print(f"user_score: {user_score}")
            else:
                should_continue = False
                print(f"user_cards: {user_cards}")
                print(f"user_score: {user_score}")
        else:
            should_continue = False
            print(f"Your final hand: {user_cards}, final score: {user_score}")
        print(f"user_cards: {user_cards}")
        print(f"user_score: {user_score}")
        while computer_score < 16:
            computer_cards.append(deal_card(cards))
            computer_score = calculate_score(computer_cards)
        else:
            print(f"computer_cards: {computer_cards}")
            print(f"computer_score: {computer_score}")
            winner = check_winner(user_cards, computer_cards)
            print(winner)
    else:
        should_continue = False
        print("Goodbye")
```
